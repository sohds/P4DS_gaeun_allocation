```python
# import library
import pandas as pd
import datetime as dt
```

```python
#데이터 불러오기
weather_data = pd.read_excel('nowon_climate_data_220302_221231.xlsx')
weather_data
```

### 전처리 진행

##### 날씨 데이터의 '일시' 열에 날짜와 시간이 합쳐져 있어 주문건수 데이터와 'time'을 기준으로 join하기 위한 전처리 필요

```python
# datetime 모듈을 활용하여 '일시' 열의 날짜와 시간을 각각 추출한 후 'time', '일자'를 열로 추가
weather_data['time'] = weather_data['일시'].dt.hour
weather_data['일자'] = weather_data['일시'].dt.date

# 분석에 필요하지 않은 '지점', '지점명', '일시' 열 삭제
weather_data = weather_data.drop('지점', axis=1)
weather_data = weather_data.drop('지점명', axis=1)
weather_data = weather_data.drop('일시', axis=1)
weather_data = weather_data.drop('풍향(deg)', axis=1) # 날씨 요소 중 구매에 큰 영향을 주지 않는 '풍향' 열 삭제

# 결과 확인
weather_data
```

```python
# 전처리된 날씨 데이터를 엑셀 파일로 저장
weather_data.to_excel('data_weather.xlsx', index=False)
```
