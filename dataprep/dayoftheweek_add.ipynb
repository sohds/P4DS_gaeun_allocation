### 메뉴 컬럼 모두 삭제 후, 요일 컬럼 추가
- pandas

```python
import pandas as pd

df = pd.read_excel('../resource/230211_1hour_data.xlsx')
```

```python
# 메뉴 컬럼 모두 삭제
df = df.iloc[:, :6]
df
```

```python
# '일자' column을 object 형식에서 datetime 형식으로 변환
df['일자'] = pd.to_datetime(df['일자'])
```

```python
pd.to_datetime(df['일자'])
```

```python
# column의 날짜에서 요일 추출
df['일자'].dt.dayofweek
```

```python
df['day of the week'] = df['일자'].dt.dayofweek
df
```

```python
# 제대로 추가됐는지 확인
df.loc[[400, 500, 600]]
```

```python
# 엑셀 파일로 변환
df.to_excel('../resource/dayoftheweek_add.xlsx', index=False)   # index 없이 저장
```
