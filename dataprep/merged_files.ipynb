### 날짜별로 하나씩 되어있던 xlsx 파일 하나로 병합

```python
directory = '../resource/rawdata/i03_i12/'  # 파일이 있는 디렉토리
output_file = '../resource/merged_files.csv'  # 저장할 CSV 파일 이름

# 파일 목록 가져오기
files = os.listdir(directory)
xlsx_files = [file for file in files if file.endswith('.xlsx')]  # xlsx 파일만 선택

# 진행률 표시 준비
progress_bar = tqdm(total=len(xlsx_files), desc='Processing', position=0)

# 각 파일의 데이터를 읽어들여 리스트에 저장
data_frames = []
for file in xlsx_files:
    file_path = os.path.join(directory, file)  # 파일 경로 생성
    data = pd.read_excel(file_path)  # Excel 파일 읽기
    
    # 문자열 열에서 쉼표 제거
    for col in data.select_dtypes(include=['object']):  # 문자열 열만 선택
        data[col] = data[col].str.replace(',', '')  # 쉼표 제거
    
    data_frames.append(data)  # 리스트에 추가
    
    # 진행 상황 업데이트
    progress_bar.update(1)

# 리스트의 모든 데이터 프레임을 병합
all_data = pd.concat(data_frames, ignore_index=True)

# 데이터를 CSV 파일로 저장
all_data.to_csv(output_file, index=False)

progress_bar.close()  # 진행률 표시 종료
print(f"모든 파일이 '{output_file}'로 병합되어 저장되었습니다.")
```
