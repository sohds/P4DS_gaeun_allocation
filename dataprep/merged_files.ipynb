### 파일 하나로 병합

```python
import pandas as pd
import os
from tqdm import tqdm  # 진행률 표시를 위한 라이브러리

directory = 'indetail'  # 파일이 있는 디렉토리
output_file = 'merged_files.xlsx'  # 저장할 Excel 파일 이름

# 파일 목록 가져오기
files = os.listdir(directory)
xlsx_files = [file for file in files if file.endswith('.xlsx')]  # xlsx 파일만 선택

# 진행률 표시 준비
progress_bar = tqdm(total=len(xlsx_files), desc='Processing', position=0)

# 모든 파일을 담을 빈 DataFrame 생성
all_data = pd.DataFrame()

# 모든 파일을 읽어들여서 DataFrame으로 합치기
for file in xlsx_files:
    file_path = os.path.join(directory, file)  # 파일 경로 생성
    data = pd.read_excel(file_path)  # Excel 파일 읽기
    all_data = all_data.append(data, ignore_index=True)  # 데이터를 기존 데이터프레임에 추가
    
    # 진행 상황 업데이트
    progress_bar.update(1)

# 데이터를 Excel 파일로 저장
all_data.to_excel(output_file, index=False)

print(f"모든 파일이 '{output_file}'로 병합되어 저장되었습니다.")
```
