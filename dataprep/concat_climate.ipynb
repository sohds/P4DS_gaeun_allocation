```python
import pandas as pd
```

```python
weather = pd.read_excel('../resource/data_weather.xlsx')
data = pd.read_excel('../resource/dayoftheweek_add.xlsx')
```

```python
# '일자'와 'time' 열을 기준으로 dayoftheweek_data와 weather_data를 left join
merged_data = pd.merge(data, weather, on=['일자', 'time'], how='left')

# 결과 확인
merged_data.head()
```

```python
final = merged_data[['일자', 'semester', 'season', 'event', 'day of the week', 'time', '기온(°C)', '풍속(m/s)', '강수량(mm)', '습도(%)', '합계']]
final.head()
```

```python
show_nan = final[final.isna( ).any(axis=1)]      # 결측치 확인 코드
show_nan
```

```python
final['강수량(mm)'] = final['강수량(mm)'].fillna(0)    # 2022년 9월 1일 비 안 온 것으로 확인, 0으로 처리

show_nan = final[final.isna( ).any(axis=1)]           # 나머지 nan 값 보기
show_nan
```

```python
# '풍속(m/s)' 열에 결측치 대체를 위한 코드
# '일자'를 기준으로 그룹화하여 '풍속(m/s)'의 평균을 계산
rainfall_mean = final.groupby('일자')['풍속(m/s)'].transform('mean')

# '강수량(mm)' 열의 결측치를 해당 날짜의 평균값으로 대체
final['풍속(m/s)'].fillna(rainfall_mean, inplace=True)
```

```python
nan_check = final.loc[[697, 1087, 1465], :]    # 풍속 nan 였던 부분 잘 들어갔는지 확인
nan_check
```

```python
final.to_excel('../resource/final_for_models.xlsx', index=False)
```
