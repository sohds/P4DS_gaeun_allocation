```python
import pandas as pd
```

```python
weather = pd.read_excel('data_weather.xlsx')
data = pd.read_excel('dayoftheweek_add.xlsx')
```

```python
# '일자'와 'time' 열을 기준으로 dayoftheweek_data와 weather_data를 left join
merged_data = pd.merge(data, weather, on=['일자', 'time'], how='left')

# 결과 확인
merged_data.head()
```
