# Visualization - Box Plot
### 주문횟수 합계 시각화를 통한 데이터의 분포 및 이상치 확인

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# 데이터 불러오기
df = pd.read_excel('../resource/dayoftheweek_add.xlsx')
df
```

## Box Plot For all semester

#### 1,2학기 전체의 주문횟수를 Box Plot으로 시각화

```python
plt.figure(figsize=(7, 6)) # 그래프의 size 설정
plt.rc('font', family='AppleGothic') # 그래프의 title font 설정
boxplot = df.boxplot(column=['합계']) # df의 '합계' 컬럼을 box plot으로 시각화
plt.title('1, 2학기 주문횟수 합계 Box Plot') # 그래프의 title 설정
plt.yticks(np.arange(0, 150, step=10))# 그래프의 y축 범위 및 간격 설정
plt.show()
```

## Box Plot For semester 1

#### 1학기의 주문횟수를 Box Plot으로 시각화

##### 1학기의 주문 데이터만 보기 위해 전처리 진행

```python
# 기존의 데이터에서 'semester' 컬럼 값이 1인 값만 추출하여 새로운 데이터프레임(semester_1) 생성
semester_1 = df.loc[df.semester == 1]
semester_1
```

#### 1학기의 주문횟수 합계를 Box Plot으로 시각화

```python
plt.figure(figsize=(7, 6)) # 그래프의 size 설정
plt.rc('font', family='AppleGothic') # 그래프의 title font 설정
boxplot = semester_1.boxplot(column=['합계']) # semester_1의 '합계' 컬럼을 box plot으로 시각화
plt.title('1학기 주문횟수 합계 Box Plot') # 그래프의 title 설정
plt.yticks(np.arange(0, 150, step=10))# 그래프의 y축 범위 및 간격 설정
plt.show()
```

## Box Plot For semester 2

#### 2학기의 주문횟수를 Box Plot으로 시각화

##### 2학기의 주문 데이터만 보기 위해 전처리 진행

```python
# 기존의 데이터에서 'semester' 컬럼 값이 2인 값만 추출하여 새로운 데이터프레임(semester_2) 생성
semester_2 = df.loc[df.semester == 2]
semester_2
```

#### 2학기의 주문횟수 합계를 Box Plot으로 시각화

```python
plt.figure(figsize=(7, 6)) # 그래프의 size 설정
plt.rc('font', family='AppleGothic') # 그래프의 title font 설정
boxplot = semester_2.boxplot(column=['합계']) # semester_2의 '합계' 컬럼을 box plot으로 시각화
plt.title('2학기 주문횟수 합계 Box Plot') # 그래프의 title 설정
plt.yticks(np.arange(0, 150, step=10))# 그래프의 y축 범위 및 간격 설정
plt.show()
```

## Box Plot For All

#### 1,2학기, 1학기, 2학기의 각 Box Plot을 한 그래프에 나타내어 데이터 분포를 한번에 확인

```python
plt.figure(figsize=(7, 6)) # 그래프의 size 설정
plt.rc('font', family='AppleGothic') # 그래프의 title font 설정
plt.boxplot([df['합계'], semester_1['합계'], semester_2['합계']], labels=['1,2학기','1학기', '2학기'])# df의 '합계' 컬럼과 semester_1의 '합계'컬럼, semester_2의 '합계'컬럼을 box plot으로 시각화
plt.title('학기별 주문횟수 합계 Box Plot') # 그래프의 title 설정
plt.yticks(np.arange(0, 150, step=10))# 그래프의 y축 범위 및 간격 설정
plt.grid() # 격자 설정
plt.savefig('png/boxplot_for_all.png', facecolor='#ffffff', bbox_inches = 'tight') # png 파일로 저장
plt.show()
```
